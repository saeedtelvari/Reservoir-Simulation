```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.001
ndt = 500
L = 1
ndx = 10



def pressure_equation(pressure, dt, ndt, L, ndx):
    dx = L/ndx
    pressure = np.zeros(shape=[ndx+1, ndt])
    pressure[0,:] = 1
    dt_dx2 = dt/(dx*dx)
    for t in range(ndt-1):
        pressure[1:-1, t+1] = pressure[1:-1, t] + dt_dx2*(pressure[:-2, t] + pressure[2:, t] - 2*pressure[1:-1, t])
    return pressure
```

```python
p = pressure_equation(pressure, dt, ndt, L, ndx=10)
```

```python
plt.figure(figsize=(10,4))
plt.ylabel('Pressure')
plt.xlabel('Length')
timestep = 0

ndx = 2
plt.plot(np.linspace(0, L, ndx+1), pressure_equation(pressure, dt, ndt, L, ndx=ndx)[:, timestep], 'o-')
# plt.legend(f"Time = {timestep*dt}")
ndx = 4
plt.plot(np.linspace(0, L, ndx+1), pressure_equation(pressure, dt, ndt, L, ndx=ndx)[:, timestep], 'o-')

ndx = 10
plt.plot(np.linspace(0, L, ndx+1), pressure_equation(pressure, dt, ndt, L, ndx=ndx)[:, timestep], 'o-')

plt.legend(['3', '5', '11'])
```
